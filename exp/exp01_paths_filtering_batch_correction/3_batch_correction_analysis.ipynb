```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = '../../data/only_healthy_data'
figure_outputs = '../../figures/exp01/3_figures'
bc = pd.read_csv(f'{data_path}/batchfix_paths_healthy_samples_pathway_counts.csv', sep=',', index_col=0, header=0)
no_bc = pd.read_csv(f'{data_path}/filtered_paths_healthy_samples_pathway_counts.csv', sep=',', index_col=0, header=0)
```

```python
# Flatten data for comparison
bc_values = bc.values.flatten()
no_bc_values = no_bc.values.flatten()

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(bc_values, no_bc_values, s=1, alpha=0.5)
#plt.title('Scatter Plot Comparing bc and no_bc Values')
plt.xlabel('Batch Corrected Values', fontweight='bold')
plt.ylabel('Pre-Batch Correction Values', fontweight='bold')
plt.savefig(f"{figure_outputs}/S3C_batch_correction_values_scatter.png")
plt.show()
```

```python
# Calculate the lib_size for both DataFrames
no_bc_sums = no_bc.sum(axis=0)
bc_sums = bc.sum(axis=0)

# Create the scatter plot
plt.scatter(bc_sums, no_bc_sums, s = 5)

# Label the axes
plt.xlabel('Library size after batch correction')
plt.ylabel('Library size before batch correction')

#plt.title('Scatter Plot of lib_size')
plt.savefig(f"{figure_outputs}/S3D_batch_correction_libsize_scatter.png")
plt.show()
```

```python
# Count zeros in each column for both DataFrames
bc_zeros = (bc == 0).sum(axis=0)  # Count zeros column-wise in bc
no_bc_zeros = (no_bc == 0).sum(axis=0)  # Count zeros column-wise in no_bc

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(bc_zeros, no_bc_zeros, s=5)
plt.xlabel('Number of zeros in samples after batch correction')
plt.ylabel('Number of zeros in samples before batch correction')
plt.savefig(f"{figure_outputs}/S3E_batch_correction_zerocounts_scatter.png")
plt.show()
```
